```python
#Import packages
import pandas as pd
import numpy as np
```

```python
#import dataset
df = pd.read_csv('World_Development_Indicators.csv')
```

```python
#panel data
df.head()
```

```python
#check data type and missing values
df.info()
```

```python
#change .. to NaN
df.replace('..',np.nan, inplace = True)
df.head()
```

```python
# missing values per years
df.isnull().sum() #missing values
```

```python
# if there are 35 Nans in a row, drop the row(series)
df.dropna(thresh=35, axis=0, inplace=True)
```

```python
# drop repeated columns
# drop the 2017 column since most of the values are missing
df.drop('2017 [YR2017]', axis=1, inplace=True)
df.head()
```

```python
#change data type from object to numeric 
col_names = ['1980 [YR1980]', '1981 [YR1981]',
       '1982 [YR1982]', '1983 [YR1983]', '1984 [YR1984]', '1985 [YR1985]',
       '1986 [YR1986]', '1987 [YR1987]', '1988 [YR1988]', '1989 [YR1989]',
       '1990 [YR1990]', '1991 [YR1991]', '1992 [YR1992]', '1993 [YR1993]',
       '1994 [YR1994]', '1995 [YR1995]', '1996 [YR1996]', '1997 [YR1997]',
       '1998 [YR1998]', '1999 [YR1999]', '2000 [YR2000]', '2001 [YR2001]',
       '2002 [YR2002]', '2003 [YR2003]', '2004 [YR2004]', '2005 [YR2005]',
       '2006 [YR2006]', '2007 [YR2007]', '2008 [YR2008]', '2009 [YR2009]',
       '2010 [YR2010]', '2011 [YR2011]', '2012 [YR2012]', '2013 [YR2013]',
       '2014 [YR2014]', '2015 [YR2015]', '2016 [YR2016]']

df[col_names] = df[col_names].apply(pd.to_numeric, errors='coerce')

#check data type 
df.dtypes.head()
```

```python
#change column names
col_names2 = ['Country Name','Country Code','Series Name','Series Code','1980', '1981',  '1982', '1983', '1984', '1985', '1986', '1987', '1988', '1989', '1990', '1991', '1992', '1993',
             '1994', '1995', '1996', '1997', '1998', '1999', '2000', '2001', '2002', '2003', '2004', '2005', '2006', '2007', '2008',
             '2009', '2010','2011', '2012', '2013', '2014', '2015', '2016']

df.columns = col_names2
```
